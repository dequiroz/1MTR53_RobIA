# 1. Introducción de Python

##¿Que es Python?
Python es uno de los lenguajes de programación más empleados en el ámbito de la robótica e inteligencia artificial.
Se define como *un lenguaje de programación interpretado, orientado a objetos, de alto nivel y con semántica dinámica*. Esto implica lo siguiente:
- **Interpredato**: Implica que el lenguaje no es compilado. Es decir, un código de python se puede ejecutar de forma directa. Esto también implica que errores no son detectados hasta el tiempo de ejecución (run-time).
- **Orientado a objetos**: Todo en python se representa por objetos, por lo que pueden usar clases y métodos. Esto permite emplear una estructura ordenada y lógica al momento de programar.
- **Alto nivel**: Permite programar empleando palabras "comunes", logrando tener un programa funcional en poco tiempo y sin mucha complejidad.
- **Semantica dinámica**: Implica que las variables son objetos, por lo que no es necesario especificar el tipo de dato. Python puede determinar esa información.

````
#Ejemplo de semantica dinámica
x = 5   #Python reconoce que x contiene un valor numerico
x = x - 2 #La clase INT permite realizar operaciones matematicas
print(f'{x=}')
x = 'Hola' #Python reconoce que x ahora contiene una cadena
x = x.upper() #el metodo upper pertenece a la clase string y se usa para cambiar una cadena a mayusculas
print(f'{x=}')
````

# 7. Comprension de listas (List Comprehension)
Este es un concepto de python que permite generar y trabajar con listas de forma facil.
Debido a que python es un lenjuage interpretado, se puede emplear un codigo iterable en la definición de una lista para generar su valor inicial. Por ejemplo, un bucle tipo for.

En el siguiente ejemplo se usa un for y se indica que los valores que toma la variable i (de 0 a 4) se empleen para generar la lista

````
a = [i for i in range(5)]
print(a)
````

Otra forma de emplear el concepto es para simplificar código que genera listas según operaciones externas, como en el siguiente ejemplo donde se genera una lista **b** mediante un bucle en el cual se itera de forma independiente por cada item de la lista data y se almacena el valor redondeado.
Por otro lado, la lista **c** se genera mediante una conprensión, resultando en el mismo valor

````
data = [100.23, 97.68, 45.07]

b = []
for value in data:
  b.append(round(value))

c = [round(value) for value in data]

print(b)
print(c)
````

Comprensiones de listas permiten tambien colocar condicionales para definir que valores son partes de la nueva lista.

````
a = [1, 15, 20, 40, 6, 34, 100]
b = [item for item in a if item > 30]
print(f'{b = }')
````

#9. Datos Mutables e Inmutables

**Es importante poder diferenciar entre variables mutables e inmutables**: La primera implica que son variables cuyo valor puede ser modificado, mientras que la segunda son variable cuyo valor queda definido al momento de asignarles un valor y no se puede ser modificado a menos de que se vuelva a reasignar
Los tipos de variables inmutables son:
- str
- int
- float
- bool
- bytes
- tuple

mientras que los datos mutables son:
- list
- set
- dict

##Ejemplo 1
En el siguiente ejemplo se observa la definicion de una tubla (objeto inmutable) y el error que se genera al intentar realizar una modificacion

````
#Ejemplo 1
# definición de una tupla
x = (1,2)
x[0] = 3 #No se puede cambiar y retorna error
````

##Ejemplo 2
En el siguiente ejemplo, se observa la creacion de 1 tupla cuyo valor se copia en otra variable **y**. Al ser una copia, su valor no se modifica al reasignarle valores distintos a la variable **x** original
Este es el comportamiento "esperado" segun otros lenguajes de programación

````
#Ejemplo 2
# definición de una tupla
x = (1,2)
y = x  #Se crea una copia de forma automatica

x = (1,2,3) #Se reasigna otra tupla a la variable x

print(x,y) #Observamos que ambas variables tienen valores distintos asignados
````

##Ejemplo 3
Aqui se intenta hacer lo mismo que en el ejemplo anterior. Sin embargo esta vez se emplean listas (objetos mutables). Debido a ello, al realizar la asignación en la segunda linea, la variable y ahora contiene una referencia a la variable x, por lo que estas dos variable apuntan hacia el mismo objeto. Debido a ello, cualquier cambio en una de ellas se ve reflejado en la otra, como se puede observar en el resultado.

````
#Ejemplo 3
# definición de una lista
x = [1,2]
y = x  #Se crea una REFERENCIA al objeto

x[0] = 100 #Se modifica el valor del primer indice de la lista
print(x,y)
````

##Ejemplo 4
En el siguiente ejemplo se puede ver el uso de una función que retorna los N números las grandes de una lista. Sin embargo, como se están empleando objetos inmutables, se puede ver que la lista original **nums** es modificada por la función. Este comportamiento puede ser no deseado y depende del programador. Lo importante es conocer que esto puede pasar para decidir si es comportamiento deseado o no.

````
def get_largest_numbers(numbers, n):
  numbers.sort()

  return numbers[-n:]

nums = [2,3,4,1,34,123,321,1]

print(f'{nums = }')
largest = get_largest_numbers(nums, 2)
print(f'{largest = }')
print(f'{nums = }')
````

## Restricciones en arreglos

1. Todos los datos en un arreglo deben de ser del mismo tipo
2. La longitud de un arreglo no puede cambiar


````
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
print(f'el arreglo contiene solo valores numericos enteros')
print(arr)

print(f'\nel arreglo contiene solo valores numericos decimales')
arr = np.array([1, 2, 3, 4, 5, 6.0])
print(arr)


print(f'\nel arreglo contiene solo caracteres')
arr = np.array([1, 2, 3, 4, 5, '6'])
print(arr)

````

3. Si se desea agregar o extender el arreglo se debe de crear uno nuevo

4. El arreglo debe de ser rectangular

````
import numpy as np

arr = np.array([[1,2,5,4],[0,4,3]])
````

## Acceso a los datos

El acceso a los datos de los arreglos es similar a listas. Se puede realizar con el índice entero en llaves []

````
import numpy as np

lista = [1, 2, 3, 4, 5, 6]
arr = np.array(lista)


print(f'{arr = }')
print(f'{arr[2] = }')

arr[2] = 999

print(f'{arr = }')
````

````
import numpy as np

matriz1 = np.array([[1,2], [3,4], [1,2]])
matriz2 = np.array([[3,1]])

print(f'matriz1 * 3 = \n {matriz1 * 3}\n')

print(f'matriz1 + 8 = \n {matriz1 + 8}\n')

print(f'matriz1 / 2 = \n {matriz1 / 2}\n')
````

ENTRE ARREGLOS
- Realiza la operación entre cada elementos del arreglo uno por uno

````
import numpy as np

array2D = np.array([[1,2,3], [4,5,6]])
multi = np.array([1,2,3])

print(f'formas deben de ser compatibles: {array2D.shape} | {multi.shape}')

array2D * multi  #NO ES PRODUCTO MATRICIAL
````

Producto matricial

````
import numpy as np

array2D = np.array([[1,2,3], [4,5,6]])
multi = np.array([1,2,3])

array2D @ multi
````
