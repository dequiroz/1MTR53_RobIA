## Generador de dataset m谩s realista

```python
def generar_dataset_realista(n_trayectorias=300, pasos=80, dt=0.1):
    trayectorias = []

    for _ in range(n_trayectorias):
        x, y = 0, 0
        v = np.random.uniform(2,5)
        ang = np.random.uniform(-0.2,0.2)

        traj = []

        for t in range(pasos):
            ang += np.random.randn()*0.02
            vx = v*np.cos(ang)
            vy = v*np.sin(ang)

            v += np.random.randn()*0.1
            v = np.clip(v, 1, 8)

            x += vx*dt
            y += vy*dt

            traj.append([x,y,vx,vy])

        trayectorias.append(np.array(traj))
    return trayectorias
```

## Visualizaci贸n

```python
trayectorias = generar_dataset_realista(n_trayectorias=300, pasos=80, dt=0.1)

plt.figure(figsize=(6,6))
for traj in trayectorias[:30]:
    plt.plot(traj[:,0], traj[:,1])
plt.axis("equal")
plt.grid()
plt.show()
```
